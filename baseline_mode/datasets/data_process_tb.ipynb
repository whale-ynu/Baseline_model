```python
action_part = action_f.iloc[0:1000000]
```

```python
action_part['user_id'] = pd.factorize(action_part['user_id'])[0]
```

```python
action_part['cat_id'] = pd.factorize(action_part['cat_id'])[0]
```

```python
action_part['item_id'] = pd.factorize(action_part['item_id'])[0]
```

```python
action_part['seller_id'] = pd.factorize(action_part['seller_id'])[0]
```

```python
action_part['brand_id'] = pd.factorize(action_part['brand_id'])[0]
```

```python
action_part
```

```python
action_part['user_id'] = action_part['user_id'].map(lambda x: x + 1)
action_part['item_id'] = action_part['item_id'].map(lambda x: x + 1)
action_part['cat_id'] = action_part['cat_id'].map(lambda x: x + 1)
action_part['seller_id'] = action_part['seller_id'].map(lambda x: x + 1)
action_part['brand_id'] = action_part['brand_id'].map(lambda x: x + 1)
```
